```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,10)
y = x **2

plt.plot(x,y)
```

```python
plt.plot(x, y,c='red', lw=5, ls='--')
```

```python
plt.figure() # dbut de la figure
plt.plot(x,y, label='carr') # contenu
plt.plot(x, x**3, label='cube')
plt.title('figure 1')
plt.xlabel('Axe X')
plt.ylabel('Axe Y')
plt.legend()
plt.savefig('figure.jpg') 
```

```python
dataset = {f"exp{i}": np.random.randn(100) for i in range(4)}
```

```python
n = 1
x = np.linspace(0,100,100)
plt.figure()
plt.title('Expriences')
for key, values in dataset.items():
    plt.subplot(4, 1, n)
    plt.plot(x, values)
    n += 1
plt.show()
```
