```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def f(x, y):
    return np.cos(x) + np.sin(y-x)
```

```python
ax = plt.axes(projection='3d')

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = f(x, y)

ax.plot_surface(x, y, z, cmap='plasma')
```

```python
plt.contour(x, y, z)
plt.colorbar()
```

```python
from scipy import optimize
```

```python
def f(params):
    x, y = params
    return np.cos(x) + np.sin(y-x) 

minimum = optimize.minimize(f, x0=[0, 0])
```

```python
def f2(params):
    x, y = params
    return -(np.cos(x) + np.sin(y-x))
maximum = optimize.minimize(f2, x0=[0, 0])
```

```python
plt.contour(x, y, z)
plt.scatter(minimum.x[0], minimum.x[1], c='r', label='minimum')
plt.scatter(maximum.x[0], maximum.x[1], c='g', label='maximum')
plt.colorbar()
plt.legend()
```
